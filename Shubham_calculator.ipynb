```python
import tkinter
from tkinter import *

win=Tk()
win.title("Simple Calculator")
win.geometry("370x555")
win.resizable(width= False, height= False)
win.configure(bg="black")


number=""

def show(value):
    global number
    number+=value
    if len(number)<=14:
        result.config(text=number)
        

        
def delete():
    global number
    length=len(number)
    x=''
    for i in range(0,length-1):
        x=x+number[i]
    number=x
    result.config(text=number)
    

def clear():
    global number
    number=""
    result.config(text=number)
    

    
def calculate():
    global number
    res=""
    if number != "":
        try:
            res=(eval(number))
        except:
            res="error"
            number=""
    result.config(text=res)


result= Label(win, text="", width=25, height=2, font=("arial",30), bd=6, bg="cadetblue", relief=RIDGE)
result.pack()


Button(win, text="AC", width=3, height=1,font=("arial",30,"bold"), bd=5, bg="#808080", fg="black", command=lambda: clear()).place(x=6,y=105)
Button(win, text="%", width=3, height=1, font=("arial",30,"bold"), bd=5, bg="#808080", fg="black", command=lambda: show("%")).place(x=98,y=105)
Button(win, text="C", width=3, height=1, font=("arial",30,"bold"), bd=5, bg="#808080", fg="black",command=lambda: delete()).place(x=188,y=105)
Button(win, text="/", width=3, height=1, font=("arial",30,"bold"), bd=5, bg="#ff6600", fg="white", command=lambda: show("/")).place(x=280,y=105)

Button(win, text="7", width=3, height=1, font=("arial",30,"bold"), bd=5, bg="#5A5A5A", fg="#ff6600", command=lambda: show("7")).place(x=6,y=195)
Button(win, text="8", width=3, height=1, font=("arial",30,"bold"), bd=5, bg="#5A5A5A", fg="#ff6600", command=lambda: show("8")).place(x=98,y=195)
Button(win, text="9", width=3, height=1, font=("arial",30,"bold"), bd=5, bg="#5A5A5A", fg="#ff6600", command=lambda: show("9")).place(x=188,y=195)
Button(win, text="*", width=3, height=1, font=("arial",30,"bold"), bd=5, bg="#ff6600", fg="white", command=lambda: show("*")).place(x=280,y=195)

Button(win, text="4", width=3, height=1, font=("arial",30,"bold"), bd=5, bg="#5A5A5A", fg="#ff6600", command=lambda: show("4")).place(x=6,y=285)
Button(win, text="5", width=3, height=1, font=("arial",30,"bold"), bd=5, bg="#5A5A5A", fg="#ff6600", command=lambda: show("5")).place(x=98,y=285)
Button(win, text="6", width=3, height=1, font=("arial",30,"bold"), bd=5, bg="#5A5A5A", fg="#ff6600", command=lambda: show("6")).place(x=188,y=285)
Button(win, text="-", width=3, height=1, font=("arial",30,"bold"), bd=5, bg="#ff6600", fg="white", command=lambda: show("-")).place(x=280,y=285)

Button(win, text="1", width=3, height=1, font=("arial",30,"bold"), bd=5, bg="#5A5A5A", fg="#ff6600", command=lambda: show("1")).place(x=6,y=375)
Button(win, text="2", width=3, height=1, font=("arial",30,"bold"), bd=5, bg="#5A5A5A", fg="#ff6600", command=lambda: show("2")).place(x=98,y=375)
Button(win, text="3", width=3, height=1, font=("arial",30,"bold"), bd=5, bg="#5A5A5A", fg="#ff6600", command=lambda: show("3")).place(x=188,y=375)
Button(win, text="+", width=3, height=1, font=("arial",30,"bold"), bd=5, bg="#ff6600", fg="white", command=lambda: show("+")).place(x=280,y=375)

Button(win, text="0", width=3, height=1, font=("arial",30,"bold"), bd=5, bg="#5A5A5A", fg="#ff6600", command=lambda: show("0")).place(x=6,y=465)
Button(win, text="00", width=3, height=1,font=("arial",30,"bold"), bd=5, bg="#5A5A5A", fg="#ff6600", command=lambda: show("00")).place(x=98,y=465)
Button(win, text=".", width=3, height=1, font=("arial",30,"bold"), bd=5, bg="#5A5A5A", fg="orange", command=lambda: show(".")).place(x=188,y=465)
Button(win, text="=", width=3, height=1, font=("arial",30,"bold"), bd=5, bg="#ff6600", fg="white", command=lambda: calculate()).place(x=280,y=465)



win.mainloop()
```
